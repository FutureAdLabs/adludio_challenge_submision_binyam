```python

import json
import pandas as pd

with open('../data/global_design_data.json') as json_file:
  nested_json = json.load(json_file)

out = {}
def flatten(x, name=''):
    if type(x) is dict:
        for a in x:
            flatten(x[a], name + a + '&')
    else:
        out[name[:-1]] = x
    return out

df = pd.Series(flatten(nested_json)).to_frame()

```

```python
df2 = df.reset_index()
df2.head()
```

```python
def restructure(df1):
    cols = [[], [], [], [], [], []]
    for i in range(len(df1)):
        lis = df1.iloc[i,0].split("&")
        for k in range(6):
            try:
                cols[k].append(lis[k])
            except:
                cols[k].append("NAv")
    df_new = pd.DataFrame()

    df_new["game_key"] = cols[0]
    df_new[f"request_id"] = cols[1]
    df_new[f"feature"] = cols[2]
    df_new[f"feature_type"] = cols[3]
    df_new[f"feature_variety"] = cols[4]
    df_new[f"sub_feature"] = cols[5]
    df_new[f"feature_value"] = df1.iloc[:,1].tolist()
    return df_new
```
