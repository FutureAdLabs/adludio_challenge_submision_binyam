```python
def generate_word_cloud(col, title):

    text = " ".join(str(item) for item in col)
    word_cloud = WordCloud(width=600, 
                height=600, collocations=False, 
                background_color="white", 
                stopwords=STOPWORDS).generate(text)
    plt.figure(figsize=(20, 12))
    plt.imshow(word_cloud, interpolation="bilinear")
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.show()

```

```python
# checking description words
col = util.clean_text(data.sample(1000)["descriptions"])
generate_word_cloud(col, "placeholder")
```

```python
# checking countries
col = util.clean_text(data["serving_location"])
generate_word_cloud(col, "placeholder")
```

```python
# checking the repeating values in text design features.
col = util.clean_text(data[((data.design_feature == "text") & (data.types == "impression"))]["feature_value"])
generate_word_cloud(col, "text feature values")
```

```python
# checking the repeating values in text design features.
col = util.clean_text(data[((data.design_feature == "text") & (data.types == "first_dropped"))]["feature_value"])
generate_word_cloud(col, "text feature values")
```

```python
# checking the repeating values in text design features.
col = util.clean_text(data[((data.design_feature == "labels") & (data.types == "impression"))]["feature_value"])
generate_word_cloud(col, "text feature values")
```

```python
# checking the repeating values in text design features.
col = util.clean_text(data[((data.design_feature == "labels") & (data.types == "first_dropped"))]["feature_value"])
generate_word_cloud(col, "text feature values")
```

```python
# time series plot
data_time = data[data.browser_ts > '2020/06/27 23:59:59.999']
data_time = data_time.set_index("browser_ts")["volume_agreed"]
plt.figure(figsize=(20, 6))
plt.title("volume trend")
data_time.plot()
plt.show()
```

```python
data_time = data[data.browser_ts > '2020/06/27 23:59:59.999']
data_time = data_time.set_index("browser_ts")["gross_cost_or_budget"]
plt.figure(figsize=(20, 6))
plt.title("budget trend")
data_time.plot()
plt.show()
```

```python
# ch3ecking relationship etween agreed volme and cost
data_scat = data[data.volume_agreed < 1500000]
plt.figure(figsize=(8, 6))
plt.title("budget trend")
plt.xlabel("budget")
plt.ylabel("volume")
plt.scatter(data_scat["volume_agreed"], data_scat["gross_cost_or_budget"])
```

### Observations
- As can be seen from the above table, chrome mobile has the largest percentage for all three categories.
- Before 2021 around 10nth month, the agreed volume of budget was relatively stable and low, but after sep, 2021 it dramatically increased
